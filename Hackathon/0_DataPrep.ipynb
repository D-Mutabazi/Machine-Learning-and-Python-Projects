```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_og = pd.read_csv('data/billing_data.csv', index_col='usage_end_time', parse_dates=True)
```

```python
df = df_og.copy()

sum_df = df.groupby(df.index).count()

sum_df
```

```python
df_service = df.groupby('service_type').count()

plt.bar(df_service.index, df_service['cost'])
plt.xticks(rotation = 90)
plt.show()
```

```python
# Based on the usage unit, what are the boxplots of the cost

plt.figure(figsize=(10, 6))
sns.boxplot(x='usage_unit', y='cost', data=df)
plt.title('Boxplot of Cost by Usage Unit')
plt.xlabel('Usage Unit')
plt.ylabel('Cost (USD)')
plt.yscale('log')  # Optional: If there are outliers or a wide range, use log scale for better visualization
plt.show()

```
